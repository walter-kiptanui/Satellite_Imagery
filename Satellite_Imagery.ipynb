```python

# Step 1: Provide Google Drive file IDs for the zipped datasets
file_ids = {
    "Kenya": "1CyQ-xz0pPz5NPSfn1oo3R26KwSfYAjQv",  # File ID for Kenya
    "Tanzania": "1e8276ygPS_ZxkecAIn6Bzvlj7lULsYdM",  #File ID for Tanzania
    "Uganda": "1oS0e7XoY9F_LIfZXVEJRSgvADA6bYtoM"   #File ID for Uganda
}

# Step 2: Download the zipped files
for country, file_id in file_ids.items():
    zip_file = f"{country}.csv.gz"
    gdown.download(f"https://drive.google.com/uc?id={file_id}", zip_file, quiet=False)

    # Step 3: Extract the contents
    with zipfile.ZipFile(zip_file, 'r') as zip_ref:
        zip_ref.extractall(f"./{country}")

# Step 4: Load the extracted datasets into DataFrames
dataframes = {}
for country in file_ids.keys():
    # Assuming the CSV file is named "data.csv" inside each zip
    file_path = os.path.join(country, "data.csv")
    dataframes[country] = pd.read_csv(file_path)

# Step 5: Combine the DataFrames if needed
combined_df = pd.concat(dataframes.values(), ignore_index=True)

# Step 6: Display or save the combined DataFrame
print(combined_df.head())
combined_df.to_csv("combined_countries.csv", index=False)
```

```python

# Example capital cities dataset
capital_cities = pd.DataFrame({
    'Country': ['Country1', 'Country2', 'Country3'],
    'Capital': ['Capital1', 'Capital2', 'Capital3'],
    'Latitude': [lat1, lat2, lat3],
    'Longitude': [lon1, lon2, lon3]
})

# Example building data (GeoJSON or CSV)
buildings = gpd.read_file("buildings.geojson")

# Convert buildings to centroids
buildings['centroid'] = buildings.geometry.centroid
buildings['lat'] = buildings['centroid'].y
buildings['lon'] = buildings['centroid'].x

# Radius calculation
def count_buildings_near_capital(lat, lon, radius_km):
    capital_coords = (lat, lon)
    buildings['distance_km'] = buildings.apply(
        lambda row: haversine(capital_coords, (row['lat'], row['lon']), unit=Unit.KILOMETERS),
        axis=1
    )
    return (buildings['distance_km'] <= radius_km).sum()

# Calculate for each capital city
radius_km = 100
for _, row in capital_cities.iterrows():
    count = count_buildings_near_capital(row['Latitude'], row['Longitude'], radius_km)
    print(f"Buildings near {row['Capital']} ({row['Country']}): {count}")
```
